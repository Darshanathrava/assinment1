```python
import pandas as pd 
import numpy as np
import random

first_names = ['jon','jane','michael','emily','william','olivia','james','sophia','robert','Ava']
last_names = ['simth','johnson','brown','lee','davis','wilson','anderson','harris','clarl','taylor']

num_samples = 400

names = [random.choice(first_names) + '' + random.choice(last_names) for _ in range(num_samples)]

np.random.seed(0) 
age = np.random.randint(29, 77, num_samples)
gender = np.random.randint(0, 2, num_samples) # 0 for female, 1 for male
chest_pain = np.random.randint(0, 4, num_samples) # 0-3 types of chest pain
blood_pressure = np.random.randint(90, 200, num_samples) 
cholesterol = np.random.randint(120, 320, num_samples)
blood_sugar = np.random.randint(0, 2, num_samples) # 0 for normal, 1 for high sugsr
ecg_results = np.random.randint(0, 3, num_samples) # 0-2 different ECG results
heart_disease = np.random.randint(0, 2, num_samples) # 0 for no disease, 1 for presence of heart disease

data = pd.DataFrame({'Name':names,'Age':age,'Gender':gender,'chestpaintype':chest_pain,'bloodpressure':blood_pressure,'Cholesterol':cholesterol,'Bloodsugar':blood_sugar,'ECGresults':ecg_results,'Heartdisease':heart_disease})

data.to_csv('heart_disease_dataset.csv', index=False)

print("synthetic heart disease dataest with ramdom names created and save as 'Heartdisease_dataset.csv'.")
```
